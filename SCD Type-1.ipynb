```python
spark.sql(f"""
create table if not exists gold_db.category (
    CategoryKey int,
    Category string,
    last_updated_date timestamp
)
location 'abfss://gold@bhaskaradlsaug.dfs.core.windows.net/category'
""")
```

```python
%sql
MERGE INTO gold_db.category AS tgt
USING silver_db.category AS src
ON tgt.CategoryKey = src.CategoryKey
WHEN MATCHED THEN
  UPDATE SET
      tgt.Category = src.Category,
      tgt.last_updated_date = current_timestamp()
WHEN NOT MATCHED THEN
  INSERT (
      CategoryKey,
      Category,
      last_updated_date
  )
  VALUES (
      src.CategoryKey,
      src.Category,
      current_timestamp()
  )
```
