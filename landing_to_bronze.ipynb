### Reading data from ADLS and writing data into ADLS with parquet format

### Reading data from ADLS landing and create delta table in bronze_db

```python
src_cal_path=f"abfss://landing@{stoarge_account_name}.dfs.core.windows.net/calendar/calendar.parquet"
trg_cal_path=f"abfss://bronze@{stoarge_account_name}.dfs.core.windows.net/calendar/"
cal_df=spark.read.parquet(src_cal_path)
cal_df.write.format("delta").mode("append").save(trg_cal_path)
# Register the Delta table in bronze_db
spark.sql(f"""
    CREATE TABLE IF NOT EXISTS bronze_db.calendar
    USING DELTA 
    LOCATION '{trg_cal_path}'
""")
```
